Problema: Elaborar um simulador de braço robótico, em duas dimensões que represente na forma cinética o posicionamento do ponto (X,Y), para ϴ1 e ϴ2 independentes e uniformes no intervalo 0 a 2$\pi$ .

---





Nesse contexto temos:

0 $\geq$ ϴ1 e ϴ2 $\geq$ 2π

ϴ1 e ϴ2 são valores aleatórios

Aplicando esses valores em código teremos:

Com base no contexto apresentado calcular:



```python
from re import A
from sympy.ntheory.residue_ntheory import randint
import matplotlib.pyplot as plt
import numpy as np
import sympy 
from sympy import *
init_printing(pretty_print=True)
in_array =[]
out_array=[]
for i in Range(3000):

  a_L1 = 1
  a_L2 = 1
  a_T1 = randint(0,360)
  a_T2 = randint(0,360)

  x1 = np.cos(a_T1)*a_L1 + np.cos(a_T1+a_T2)*a_L2
  y1 = np.sin(a_T1)*a_L1 + np.sin(a_T1+a_T2)*a_L2


  in_array.append(x1)
  out_array.append(y1)
 

plt.scatter(in_array, out_array, color = 'red', marker = "o") 
plt.title("Pontos de posicionamento da ferramenta") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.axvline(0, color='k')
plt.axhline(0, color='k')
plt.show() 

```
